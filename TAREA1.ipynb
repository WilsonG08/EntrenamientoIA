```python
# Importar la libreria para la regresion lineal
from sklearn.linear_model import LinearRegression
#Funcion para dividir datos y prueba
from sklearn.model_selection import train_test_split
# Importar funcion de error cuadratico medio
from sklearn.metrics import mean_squared_error,r2_score
#grafico
import matplotlib.pyplot as plt


import numpy as np
import pandas as pd
```

```python
# Importar los datos
datosCSV = pd.read_csv("seguro_salud.csv")



# Datos de entrada (caracteristicas)
x = datosCSV.iloc[:, :-1] # Todas las columnas excepto la ultima
# Datos de salida (caracteristicas)
y = datosCSV.iloc[:, -1] # Ultima columna



# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=25)



# Crear el modelo de regresión lineal (algoritmo)
modelLineal = LinearRegression()



# Entrenar el modelo con los datos de entrenamiento
modelLineal.fit(X_train, y_train)



# Probar el modelo con los datos de prueba
y_obt=modelLineal.predict(x_test)



#Evaluar el modelo con las métricas 
#Error cuadratico medio
error= mean_squared_error(y_test,y_obt)
print("============================================")
print("El error cuadratico medio es: ",error)

#Coeficiente de determinacion
r2= r2_score(y_test,y_obt)
print("El coeficiente de determinacion es: ",r2)

# Imprimir los coeficientes de la regresión
print('Coeficientes:', modelLineal.coef_)
print('Intercepto:', modelLineal.intercept_)
print("============================================")
```

```python
# CASO 1: Hombre, 33 años, sin hijos, fumador, con IBM de 40 kg/m
prediccion_seguro_costo1 = modelLineal.predict([[33, 0, 40, 0, 1]])[0]
print("El costo de seguro de salud para el caso 1 es :", prediccion_seguro_costo1)
print("======================================================")

# CASO 2: Mujer, 45 años, 2 hijos, fumador, con IBM de 25 kg/m
prediccion_seguro_costo2 = modelLineal.predict([[45, 1, 25, 2, 1]])[0]
print("El costo de seguro de salud para el caso 2 es :", prediccion_seguro_costo2)
print("======================================================")


# CASO 3: Hombre, 60 años, 3 hijos, no fuma, con IBM de 33 kg/m
prediccion_seguro_costo3 = modelLineal.predict([[60, 0, 33, 3, 0]])[0]
print("El costo de seguro de salud para el caso 3 es :", prediccion_seguro_costo3)
print("======================================================")
```

```python
# Gráfico
plt.scatter(x['age'], y, color ='orange')
plt.title('Relación entre la edad y el costo del seguro ')
plt.xlabel('Edad')
plt.ylabel('Cargo')
plt.legend()

# Mostrar la gráfica
plt.show()
```
